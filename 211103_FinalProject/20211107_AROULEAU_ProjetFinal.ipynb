<h2>0.3. Outils de conversion </h2>

Cette fonction va permettre de convertir les dates d'enregistrement extraites du site internet. Elles vont servir essentiellement:
- pour le téléchargement de tous les fichiers
- localiser temporellement chaque ligne appartenant à un voilier spécifique dans ledataframe

```python
def CONVERT_TIME_FORMAT (LIST_URL_Time :'YYYY-MM-DD hh:mm:ss')\
      ->'datetime.datetime(YYYY,M,D,H,M),[YYYYMMDD_HHMMSS]':
    
    LIST_Time =[]               # Daily rank date list extracted from URL_Rank
    LIST_FileID = []            # File ID list
    for i in LIST_URL_Time:     # Converting all the daily rank date formats
        LIST_Time.append(dt.datetime.strptime(i, "%Y-%m-%d %H:%M:%S"))
        LIST_FileID.append(dt.datetime.strptime(i, "%Y-%m-%d %H:%M:%S").
                           strftime("%Y%m%d_%H%M%S"))
    return LIST_Time,LIST_FileID
```

```python
def CHECK_DATA_DIR(DIR_Main:'Dir_String', TYPE_Data:'Type_String')\
      ->'BOOL_ExistDir, DIR_Data':
    
    DIR_Data = os.path.join(DIR_Main, TYPE_Data)
    if not os.path.exists(DIR_Data):   # Make subdirectory if doesn't exist
        os.mkdir(DIR_Data)
        BOOL_DirExist = False
    else:
        BOOL_DirExist = True           # Else pass
    print("The subdirectory exists "+TYPE_Data+" exists :"+\
          str(np.where(BOOL_DirExist==True,"OK","NOK")))
    return BOOL_DirExist, DIR_Data
```

><h3>L'obtention d'une liste avec toutes les dates d'enregistrement</h3>
Cette liste permet aussi d'identifier le nom et la date du fichier

```python
def GET_RANKING_DATE_LIST(URL_Ranks : 'URL_String')\
      ->'List of ranking date in AAAA-MM-DD HH:MM:SS format':
    
    
    SOUP_URLRank = BeautifulSoup(requests.get(URL_Ranks).
                                 content.decode('utf-8'))            #Get Soup
    LIST_SoupURLRank = SOUP_URLRank\
          .findAll('option')                        #Extract only options list
    LIST_RankDate = []
    for iLIST_SoupURLRank in LIST_SoupURLRank:     #Generate ranking date list
        LIST_RankDate.append(iLIST_SoupURLRank.text[2:])
    LIST_RankDate = np.unique(LIST_RankDate[1:-1])        #Delete duplications
    print("Number of daily ranking files : " + str(len(LIST_RankDate)))
    return LIST_RankDate
```

><h3>La mise à jour des fichiers une fois téléchargés</h3>
Les fichiers excel ont posé problème lors de l'ouverture avec pd.read_excel.
Nous avons besoin de passer par ce code pour y remédier.

```python
def UPDATE_FORMAT(FILE : "FileName.FileExt string",
                  name_filter : "Name filer string",
                  change : 'string')\
      ->'Nothing : Excel files are updated to be workable':
    DIR_temp = tempfile.mkdtemp()
    try:
        NAME_temp = os.path.join(DIR_temp, 'new.zip')
        with ZipFile(FILE, 'r') as r, ZipFile(NAME_temp, 'w') as w:
            for item in r.infolist():
                DATA = r.read(item.filename)           
                DATA = change(DATA)
                w.writestr(item, DATA)
        shutil.move(NAME_temp, FILE)
    finally:
        shutil.rmtree(DIR_temp)
```

><h3>Le téléchargement des fichiers Excel</h3>

```python
def DWNLD_DAILY_RANK_FILES(URL_DwnldRank : 'DownloadingURL_String',
                           LIST_FileID : 'List of string containing files ID',
                           EXT_RankFile : '.ext_file String',
                           DIR_Data : 'Data Directory string',
                           DIR_Main : 'Main Directory string')\
      -> 'Nothing : Excel files downloaded and saved in data directory':
    os.chdir(DIR_Data)                                    # Go to subdirectory
    with tqdm(range(len(LIST_FileID)),
              desc = "Downloading progress bar")\
              as outer :
        for iLIST_FileID in LIST_FileID:    # Download each Excel ranking file
            resp = requests.get(URL_DwnldRank+iLIST_FileID+EXT_RankFile)
            FILE = "vendeeglobe_"+iLIST_FileID+EXT_RankFile
            with open(FILE,'wb')as file:
                file.write(resp.content)
            UPDATE_FORMAT(FILE,name_filter='xl/styles.xml',
                          change=lambda d:re.sub(b'xxid="\d*"',b"",d))
            outer.update()
    os.chdir(DIR_Main)
```

<h1>1.1. Classement des données contenu dans Excel</h1>
Il existe deux types de données dans ces fichiers:
- Le classement des voiliers en cours de course
- Le classement final de ces voiliers, une fois leur parcours terminé

Ces deux types de données sont intégrés dans la même page Excel, et leur hauteur varie avec le temps :
- Au départ : nous avons uniquement des classement journalier, des voiliers en cours de courses
- A partir du 1er arrivé : nous avons d'abord le classement finale. En dessous, le classement des voiliers encore en cours de course
- Au dernier fichier : nous avons le classement final.

Dans l'énoncé, il est indiqué qu'il est possible de s'arrêter avant le premier classement. Pour mon cas, j'ai décidé d'étudier jusqu'au bout.

La fonction ci-dessous permet de classer ces deux types de données dans chaque fichier excel.
S'il existe des données types 'en cours de course' : alors le dataframe va cumuler des données, au fur et à mesure.
S'il existe des données de 'classement final" : seules les données du dernier fichier sera retenu.

```python
def CLASS_XLS_DATA(iLIST_FileID, iLIST_RankDate):
    FILE_Name = "vendeeglobe_"+iLIST_FileID+".xlsx"
    DF_RANK = pd.read_excel(FILE_Name)
    BOOL_RaceState= [False, False]    # Boolean vect. confirming if data in file containt [Race progressing, Race finished]
    
    if DF_RANK.loc[DF_RANK["Unnamed: 5"]=="Latitude\nLatitude"].index.values:
        BOOL_RaceState[0] = True
    if DF_RANK.loc[DF_RANK['Unnamed: 8']=="Temps de course\nRace time"].index.values:
        BOOL_RaceState[1] = True
    
    if BOOL_RaceState == [True, False]:
        idx_RACE_Prog = DF_RANK.loc[DF_RANK["Unnamed: 5"]=="Latitude\nLatitude"].index.values[0]
        DF_Daily_Rank = pd.read_excel(FILE_Name,header = idx_RACE_Prog+1,nrows=33)
        DF_Final_Rank = pd.DataFrame()

    elif BOOL_RaceState == [True,True]:
        idx_RACE_Prog = DF_RANK.loc[DF_RANK["Unnamed: 5"]=="Latitude\nLatitude"].index.values[0]
        idx_RACE_end = DF_RANK.loc[DF_RANK['Unnamed: 8']=="Temps de course\nRace time"].index.values[0]+2
        DF_Final_Rank = pd.read_excel(FILE_Name,header = idx_RACE_end, nrows=idx_RACE_Prog-idx_RACE_end-1)
        DF_Daily_Rank = pd.read_excel(FILE_Name,header = idx_RACE_Prog+1,nrows=33-(idx_RACE_Prog-idx_RACE_end-1))

    elif BOOL_RaceState == [False, True]:
        idx_RACE_end = DF_RANK.loc[DF_RANK['Unnamed: 8']=="Temps de course\nRace time"].index.values[0]+2
        DF_Final_Rank = pd.read_excel(FILE_Name,header = idx_RACE_end, nrows=33)
        DF_Daily_Rank = pd.DataFrame()
        
        
        
    return DF_Daily_Rank, DF_Final_Rank
```

```python
def GEN_RANKING_DF(LIST_FileID, LIST_RankDate,DIR_Data,DIR_Main):
    os.chdir(DIR_Data)
    print ("Loading daily rank file and generating DataFrames")
    with tqdm(range(len(LIST_FileID)),
                  desc = "Loading progress bar")\
                  as outer :
        DF_DailyRank = pd.DataFrame()
        for iDailyRank in range (len(LIST_FileID)):    # Download each Excel ranking file
            iDF_Daily_Rank, DF_Final_Rank = CLASS_XLS_DATA(LIST_FileID[iDailyRank], LIST_RankDate[iDailyRank])
            if iDailyRank < len(LIST_FileID)-1:
                iDF_Daily_Rank['Date_Time']=LIST_RankDate[iDailyRank]
            DF_DailyRank = DF_DailyRank.append(iDF_Daily_Rank, ignore_index = True)
            outer.update()
    print ("Generation of Ranking DataFrames finished")
    os.chdir(DIR_Main)
    return DF_DailyRank, DF_Final_Rank
```

><h2>2.2. Nettoyage et mise en forme des données du Dataframe Classement final</h2>

>><h3> 2.2.1. Conversion des string en format temporel de durée (Timedelta)</h3>
Le code ci-dessous permet de convertir string exprimant les durées de courses ou les écarts entre voiliers, en format numérique temporel.

```python
def CONVERT_DURATION_STR_TO_DT(STR_Duration):
    STR_Duration = STR_Duration.split(' ')
    if len(STR_Duration)==4:
        [j,h,m,s]=[string.rstrip('jhmins') for string in STR_Duration]
    if len(STR_Duration)==3:
        [j,h,m,s]=[0] + [string.rstrip('jhmins') for string in STR_Duration]
    if len(STR_Duration)==2:
        [j,h,m,s]=[0,0] + [string.rstrip('jhmins') for string in STR_Duration]
    DT_Duration = dt.timedelta(days= int(j), hours=int(h), minutes=int(m), seconds=int(s))
    return DT_Duration
```

>><h3>2.2.2. Script de nettoyage et de mise en forme des données du dataframe de classement final</h3>

```python
def CLEAN_FINAL_RANK_DF(DF_Final_Rank):
    DF_Final_Rank = DF_Final_Rank.dropna(axis=1,how='all')
    DF_Final_Rank=DF_Final_Rank[DF_Final_Rank["Unnamed: 1"]!='RET']
    DF_Final_Rank=DF_Final_Rank.rename(columns={'Unnamed: 1' : 'Final Rank',
                                                'Unnamed: 2' : 'Sail_ID',
                                                'Unnamed: 3' : 'Name',
                                                'Unnamed: 7' : 'Arrival',
                                                'Unnamed: 8' : 'Race Duration',
                                                'Unnamed: 13' : 'Gap Leader',
                                                'Unnamed: 15': 'Gap Previous',
                                                'Vitesse\nSpeed':'Ortho Speed [kts]',
                                                'Distance\nDistance':'Ortho Dist [nm]',
                                                'Unnamed: 18':'Distance Ratio',
                                                'Vitesse\nSpeed.1':'Ground Speed [kts]',
                                                'Distance\nDistance.1':'Ground Dist [nm]'})

    DF_Final_Rank[('Final Rank')]=DF_Final_Rank['Final Rank'].map(lambda x:x.split('\n')[0]).astype(int)
    DF_Final_Rank.Sail_ID=DF_Final_Rank.Sail_ID.map(lambda x: x.lstrip('\n').split('\n')[-1]).astype(str)
    DF_Final_Rank['SailName']=DF_Final_Rank.Name.map(lambda x:x.split('\n')[1])
    DF_Final_Rank.Arrival=DF_Final_Rank.Arrival.map(lambda x:x.rstrip('FR'))
    for iMainCol in ['Name','Race Duration']:
        DF_Final_Rank[iMainCol]=DF_Final_Rank[iMainCol].map(lambda x:x.split('\n')[0])
        if iMainCol == 'Name':
            DF_Final_Rank.Name=DF_Final_Rank.Name.map(lambda x:" ".join([x.split('\n')[0].split(' ',1)[0],
                                                                         x.split('\n')[0].split(' ',1)[1].upper()]))
    for iMainCol in ['Ortho Speed [kts]','Ortho Dist [nm]','Distance Ratio','Ground Speed [kts]','Ground Dist [nm]']:
        DF_Final_Rank[iMainCol]=DF_Final_Rank[iMainCol].map(lambda x: x.rstrip('FR%ktsnm')).astype(float)
    DF_Final_Rank.Arrival = [dt.datetime.strptime(time, "%d/%m/%Y %H:%M:%S ") for time in DF_Final_Rank.Arrival]
    DF_Final_Rank[['Gap Leader','Gap Previous']] = DF_Final_Rank[['Gap Leader','Gap Previous']].fillna('00h 00min 00s')
    for iMainCol in ['Race Duration','Gap Leader','Gap Previous']:
        DF_Final_Rank[iMainCol]=DF_Final_Rank[iMainCol].map(lambda x: CONVERT_DURATION_STR_TO_DT(x))
    DF_Final_Rank = DF_Final_Rank.reindex(columns=['Sail_ID','Final Rank', 'Name','Arrival','Race Duration',
                                                   'Gap Leader','Gap Previous','Ortho Speed [kts]','Ortho Dist [nm]',
                                                   'Distance Ratio','Ground Speed [kts]','Ground Dist [nm]','SailName'])
    DF_Final_Rank.set_index("Sail_ID", inplace=True)
    return DF_Final_Rank
```

><h3>2.3.1. Extraction des données et génération d'un DataFrame contenant des techniques des voiliers par internet</h3>
>Le script ci-dessous extraits les données spécifiques de chaque voilier, tel:
>- son numéro de voile
>- le nom du voilier
>- le nom du skipper
>- ses données techniques dont ses mesures, son poids, le matériau...

```python
def GEN_SEPCS_DF (URL_Skipper):
    # Paramètres des types de données à extraire
    PARAM_Data = {
        "BoatSpecs" : ['ul', "class", "boats-list__popup-specs-list"],
        "SailName" : ['h3',"class","boats-list__boat-name"],
        "Name" : ["span","class", "boats-list__skipper-name"]
    }
    
    # Fonction d'extraction de données techniques spécifiques
    #(BoatSpecs, BoatName, SkipperName) sous forme de list
    def LIST_SOUP (URL_Skipper, PARAM_Data,iPARAM_Data):
        return BeautifulSoup(requests.get(URL_Skipper).content\
                                 .decode("utf-8")).findAll(
                                    PARAM_Data[iPARAM_Data][0],
                                    {PARAM_Data[iPARAM_Data][1]:
                                     PARAM_Data[iPARAM_Data][2]})
    # Création d'un DF Spécifications techniques du voilier vierge
    DF_TechSpecs=pd.DataFrame()
    
    # Remplissage du DF
    for iPARAM_Data in PARAM_Data:
        LIST_Soup = LIST_SOUP (URL_Skipper, PARAM_Data, iPARAM_Data)
        # Remplir d'abord les données techniques du voilier
        #(Num. voile, Anciens nom, Architece, Chantier, Date lanc., Longueur,...)
        if iPARAM_Data == list(PARAM_Data.keys())[0]:
            idx = 0
            for iLIST_Soup in LIST_Soup:
                iDATA,iVAL = np.transpose([string.split(' : ',1) for string
                                           in iLIST_Soup.text.split("\n")[1:-1]])
                DF_TechSpecs = pd.concat([DF_TechSpecs,
                                          pd.DataFrame([iVAL],columns = iDATA)],
                                             ignore_index=True)
                idx+=1
        # Ajouter ensuite les colonnes Nom de voilier, et nom du marin;
        #puis les mettre en premières colonnes
        else:
            DF_TechSpecs[iPARAM_Data] = [string.text for string in LIST_Soup]
            DF_TechSpecs = DF_TechSpecs[DF_TechSpecs.columns.tolist()[-1:]+
                                        DF_TechSpecs.columns.tolist()[:-1]]
    

    return DF_TechSpecs
```

><h3>2.3.2. Préparation des liens entre les Dataframes</h3>
Les numéros de voiles, des skippeurs ne sont pas homogènes. Des données sont également manquants.
Le script ci-dessous permet de faire le rapprochement entre :
- le numéro des voiliers dans le dataframe des classements
- le numéro des voiliers dans le dataframe données techniques

```python
def LIST_DFs_Relation (DF1, DF2):
    MAIN_List = np.sort(DF1.index.unique())
    ToReplaceList = {}
    i=0
    for iMAIN_List in MAIN_List:
        if iMAIN_List in np.sort(DF2['Sail_ID'].unique()):
            pass
        else:
            if ''.join(iMAIN_List.split(' ')) in np.sort(DF2['Sail_ID'].unique()):
                StrToChange = ''.join(iMAIN_List.split(' '))
            elif iMAIN_List.split(' ')[1] in np.sort(DF2['Sail_ID'].unique()):
                StrToChange =iMAIN_List.split(' ')[1]
            elif '0'+iMAIN_List[-1] in np.sort(DF2['Sail_ID'].unique()):
                StrToChange='0'+iMAIN_List[-1]
            elif iMAIN_List == 'FRA 01':
                StrToChange = '001'
            elif iMAIN_List == 'FRA 02':
                StrToChange = '2'
            elif iMAIN_List=='MON 10':
                StrToChange = '16'
            elif iMAIN_List=='SUI 7':
                StrToChange='SUI07'
            elif iMAIN_List=='FRA 59':
                StrToChange='nan'
            elif iMAIN_List=='GBR 777':
                StrToChange='GBR 77'
            else :
                i+=1
                print ("Nombre d'incohérence restant : " + str(i) +"\n\n")
                print (iMAIN_List+ ' : FALSE')
            ToReplaceList[StrToChange] = iMAIN_List
    print ('Table de correspondance entre les deux DF :')
    print(ToReplaceList)
    return ToReplaceList
```
